```python
suv=1000
sedan=500
hatchback=250
deposit=500 #d is for deposit
car=input("Enter the type of car you want to rent: ")
days=int(input("How many days do u want to rent the car for: "))
p=0
if car=="suv":
  p=p+suv*days
elif car=="sedan":
  p=p+sedan*days
elif car=="hatchback":
  p=p+hatchback*days
else:
    print("Invalid car type entered.")
    exit()
d=p
if days>=10:
  d=d*0.1
  print("Discount applied of Rs "+str(d))
  #print("Payment to be done: ",str(p-d))
else:
  print("No discount")
  #print("Payment to be done: ",str(p))



print("\n----- Rental Bill -----")
print(f"Car Type: {car.capitalize()}")
print(f"Days Rented: {days}")
print(f"Base Price: Rs{p}")
if d > 0:
    print(f"Discount: Rs {d}")
    print(f"Deposit: +Rs {deposit}")
    print("------------------------")
    print(f"Total Payment: Rs {p-d+deposit}")
    print("------------------------")
else:
    print("Discount: N/A (Rental period less than 10 days)")
    print(f"Deposit: +Rs {deposit}")
    print("------------------------")
    print(f"Total Payment: Rs {p+deposit}")
    print("------------------------")
```

```python

suv_price = 1000
sedan_price = 500
hatchback_price = 250
deposit_per_car = 500


rental_schedule = {
    "Monday": "SUV",
    "Tuesday": "Sedan",
    "Wednesday": "SUV",
    "Thursday": "Sedan",
    "Friday": "SUV",
    "Saturday": "Sedan",
    "Sunday": "Hatchback"
}


weeks = 4


total_cost = 0
total_suv_days = 0
total_sedan_days = 0
total_days = weeks * len(rental_schedule)


for _ in range(weeks):
    for day, car_type in rental_schedule.items():
        if car_type == "SUV":
            total_cost += suv_price
            total_suv_days += 1
        elif car_type == "Sedan":
            total_cost += sedan_price
            total_sedan_days += 1
        elif car_type == "Hatchback":
            total_cost += hatchback_price

discount = 0
if total_suv_days + total_sedan_days > 10:
    discount = (total_suv_days * suv_price + total_sedan_days * sedan_price) * 0.1

# Calculate final payment
num_cars = weeks * len(rental_schedule)  # One deposit per car per week
total_payment = total_cost - discount + num_cars * deposit_per_car

# Print detailed bill
print("\n----- Rental Bill -----")
print(f"Total Cost (before discount and deposit): Rs {total_cost}")
if discount > 0:
    print(f"Discount: -Rs {discount}")
print(f"Deposit (Rs {deposit_per_car} per car): Rs {num_cars * deposit_per_car}")
print(f"Total Payment: Rs {total_payment}")
print("------------------------")
```
