**Exp 1: Data Frame Manipulation and Filtering in Pandas**


---

In this experiment, we are manipulating and filtering a dataset of employees, their departments, salaries, and years of experience. We will perform several operations, including adding new columns, filtering the DataFrame based on multiple conditions, sorting, and grouping the data.

```python
import pandas as pd
import numpy as np

# Step 1: Create a DataFrame with employee data
data = {
    'Employee': ['John', 'Sarah', 'Paul', 'Nina', 'Tom', 'Emma', 'Robert', 'Sophia'],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'Finance', 'IT', 'HR', 'Marketing'],
    'Salary': [50000, 60000, 55000, 48000, 62000, 53000, 47000, 52000],
    'Experience': [5, 7, 3, 2, 10, 4, 1, 6],
    'Bonus': [2000, 2500, 1800, 1500, 3000, 1700, 1200, 2000]
}
df = pd.DataFrame(data)

# Step 2: Filter employees with more than 5 years of experience
filtered_experience = df[df['Experience'] > 5]

# Step 3: Filter employees earning more than 50,000 in Salary and bonus combined
df['Total_Compensation'] = df['Salary'] + df['Bonus']
filtered_salary = df[df['Total_Compensation'] > 52000]

# Step 4: Select only 'Employee' and 'Department' columns from filtered data
filtered_selection = filtered_salary[['Employee', 'Department']]

# Step 5: Sort the DataFrame by 'Experience' in descending order
sorted_df = df.sort_values(by='Experience', ascending=False)

# Step 6: Group by 'Department' and calculate the average Salary and Experience
grouped_df = df.groupby('Department').agg({'Salary': 'mean', 'Experience': 'mean'}).reset_index()

# Output results
print("Employees with more than 5 years of experience:")
print(filtered_experience)
print("\nEmployees with total compensation greater than 52,000:")
print(filtered_salary)
print("\nSelected columns (Employee and Department):")
print(filtered_selection)
print("\nSorted by Experience:")
print(sorted_df)
print("\nAverage Salary and Experience per Department:")
print(grouped_df)
```

**Exp 2: Handling Missing Data and Normalization in Pandas**


---


In this experiment, we work with a dataset of student scores in Math and Science that contains some missing values. We will handle these missing values, normalize the data using two different methods, and also perform basic descriptive statistics.

```python
# Step 1: Create a DataFrame with missing values
data = {
    'Student': ['Anvay', 'Bunty', 'Chirag', 'Dhruv', 'Eric', 'Frank'],
    'Math_Score': [85, np.nan, 78, 92, np.nan, 88],
    'Science_Score': [75, 88, np.nan, 85, 80, np.nan],
    'English_Score': [88, 90, 85, np.nan, 87, 89]
}
df = pd.DataFrame(data)

# Step 2: Handling missing values without inplace
df['Math_Score'] = df['Math_Score'].fillna(df['Math_Score'].mean())
df['Science_Score'] = df['Science_Score'].fillna(df['Science_Score'].mean())
df['English_Score'] = df['English_Score'].fillna(df['English_Score'].median())

# Step 3: Normalization using Min-Max scaling
df['Math_Score_MinMax'] = (df['Math_Score'] - df['Math_Score'].min()) / (df['Math_Score'].max() - df['Math_Score'].min())
df['Science_Score_MinMax'] = (df['Science_Score'] - df['Science_Score'].min()) / (df['Science_Score'].max() - df['Science_Score'].min())

# Step 4: Normalization using Z-score scaling
df['Math_Score_Zscore'] = (df['Math_Score'] - df['Math_Score'].mean()) / df['Math_Score'].std()
df['Science_Score_Zscore'] = (df['Science_Score'] - df['Science_Score'].mean()) / df['Science_Score'].std()

# Step 5: Descriptive statistics
summary_stats = df.describe()

# Output the DataFrame and summary statistics
print("Data after handling missing values and normalization:")
print(df)
print("\nDescriptive statistics of the dataset:")
print(summary_stats)
```

**Exp 3: Array Operations and Manipulations in NumPy**



---

In this experiment, we'll perform various array operations and manipulations using NumPy. The dataset will consist of temperatures recorded in 7 different cities over a week. We'll create an array, perform reshaping, slicing, aggregation, and filtering operations.

```python
# Step 1: Creating a NumPy array representing temperatures of 7 cities over 7 days (in Celsius)
temperatures = np.array([
    [29, 31, 32, 33, 28, 27, 30],  # City 1
    [25, 26, 27, 28, 24, 22, 26],  # City 2
    [20, 21, 19, 22, 23, 18, 20],  # City 3
    [34, 35, 36, 37, 33, 32, 34],  # City 4
    [30, 31, 29, 30, 28, 27, 31],  # City 5
    [27, 28, 26, 27, 25, 23, 27],  # City 6
    [21, 22, 20, 23, 24, 19, 21],  # City 7
])

# Step 2: Basic Array Operations
print("Original temperature data (7 cities over 7 days):")
print(temperatures)

# Transpose of the array (rows become columns, and vice versa)
transpose_temps = temperatures.T
print("\nTransposed temperature data:")
print(transpose_temps)

# Step 3: Aggregation Operations
# Mean temperature across all cities for the week
mean_temp = np.mean(temperatures)
print("\nMean temperature for all cities over the week:", mean_temp)

# Maximum temperature recorded for each city
max_temp_per_city = np.max(temperatures, axis=1)
print("\nMaximum temperature for each city:", max_temp_per_city)

# Minimum temperature recorded for each day across all cities
min_temp_per_day = np.min(temperatures, axis=0)
print("\nMinimum temperature across all cities for each day:", min_temp_per_day)

# Step 4: Slicing and Filtering Operations
# Get temperatures for City 3 and City 5
city_3_5_temps = temperatures[[2, 4], :]
print("\nTemperatures for City 3 and City 5:")
print(city_3_5_temps)

# Filter days where temperature for City 1 was greater than 30 degrees
city_1_high_temps = temperatures[0, temperatures[0] > 30]
print("\nDays when City 1 had temperatures greater than 30 degrees:")
print(city_1_high_temps)

# Step 5: Conditional Filtering
# Get all temperatures where any city had temperatures above 33 degrees
above_33 = temperatures[temperatures > 33]
print("\nAll temperatures greater than 33 degrees across all cities:")
print(above_33)

# Step 6: Reshaping the array (concise display)
# Reshape the 7x7 array into a 49x1 array, showing only the first 10 elements
reshaped_array = temperatures.reshape(49, 1)
print("\nReshaped array (showing first 10 values):")
print(reshaped_array[:10])
```

**Exp 4: Feature Engineering and Visualization in Pandas**


---

In this experiment, we will demonstrate feature engineering techniques in Pandas using a dataset of cars. The dataset includes attributes such as car type, engine size, horsepower, and price. We will create new features, apply transformations, and visualize the results using Matplotlib

```python
import matplotlib.pyplot as plt

# Step 1: Creating a dataset of cars
data = {
    'Car_Type': ['SUV', 'Sedan', 'Hatchback', 'SUV', 'Sedan', 'Hatchback', 'SUV'],
    'Engine_Size': [3.5, 2.0, 1.6, 4.0, 1.8, 1.2, 3.2],  # in liters
    'Horsepower': [290, 160, 130, 350, 180, 105, 310],
    'Price': [40000, 20000, 18000, 50000, 22000, 15000, 42000]  # in USD
}

df = pd.DataFrame(data)

# Step 2: Feature Engineering - Creating new features
# Creating a new feature: Price per Horsepower
df['Price_per_HP'] = df['Price'] / df['Horsepower']

# Binning the Engine Size into categories (Small, Medium, Large)
bins = [0, 2, 3.5, 5]
labels = ['Small', 'Medium', 'Large']
df['Engine_Category'] = pd.cut(df['Engine_Size'], bins=bins, labels=labels)

# Log transformation of the Price to normalize it
df['Log_Price'] = np.log(df['Price'])

# Step 3: Visualizing the data
# Scatter plot - Price vs Horsepower
plt.figure(figsize=(8, 6))
plt.scatter(df['Horsepower'], df['Price'], color='blue', label='Price vs Horsepower')
plt.title('Car Price vs Horsepower')
plt.xlabel('Horsepower')
plt.ylabel('Price (in USD)')
plt.grid(True)
plt.legend()
plt.show()

# Bar chart - Count of Cars by Engine Category
plt.figure(figsize=(8, 6))
df['Engine_Category'].value_counts().plot(kind='bar', color='orange')
plt.title('Count of Cars by Engine Category')
plt.xlabel('Engine Category')
plt.ylabel('Count of Cars')
plt.grid(True)
plt.show()

# Step 4: Display the updated DataFrame
print("Updated DataFrame with new features:")
print(df)
```

**Exp 5: Web Scraping and Data Storage**


---

In this experiment, we will perform web scraping using the Beautiful Soup library to extract data from a website. We will scrape some car data from a hypothetical webpage and then store the data in a Pandas DataFrame and a CSV file for further analysis.

```python
!pip install requests beautifulsoup4 pandas
```

```python
import requests
from bs4 import BeautifulSoup

# Step 1: URL to scrape data
url = 'https://amazon.com'

# Step 2: Make a request
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Step 3: Extract and print some data
titles = soup.find_all('h1')
for title in titles:
    print(title.get_text())

# Step4: Saving a new DataFrame to CSV
new_data = pd.DataFrame({'Title': [title.get_text() for title in titles]})
new_data.to_csv('scraped_data.csv', index=False)
```

**Exp 6: Interactive Line Plot with Plotly**


---
In this experiment, we will visualize the stock prices of three hypothetical companies over a month. This will help us analyze trends and compare the performance of these companies.


```python
import plotly.express as px

# Step 1: Generate random stock price data for three companies over a month
np.random.seed(42)  # For reproducibility
companies = ['Company A', 'Company B', 'Company C']
days = pd.date_range(start='2024-01-01', periods=30).date

# Generate random stock prices
data = {
    'Day': np.tile(days, len(companies)),
    'Company': np.repeat(companies, 30),
    'Price': np.concatenate([np.random.randn(30).cumsum() + 100 for _ in companies])
}

df_prices = pd.DataFrame(data)

# Step 2: Create the interactive line plot using Plotly Express
fig = px.line(
    df_prices,
    x='Day',
    y='Price',
    color='Company',
    title='Stock Prices of Companies Over a Month',
    labels={'Price': 'Stock Price (USD)', 'Day': 'Date'}
)

# Step 3: Show the plot
fig.show()
```

**Exp 8:  Interactive Scatter Plot with Plotly**


---

In this experiment, we will create an interactive scatter plot to visualize the relationship between the average temperature and ice cream sales over a month. This will help us understand how temperature affects ice cream sales.

```python
# Step 1: Generate random temperature and ice cream sales data
np.random.seed(42)  # For reproducibility
days = pd.date_range(start='2024-06-01', periods=30).date

# Generate random temperatures (in Celsius) and ice cream sales
temperatures = np.random.uniform(20, 40, size=len(days))  # Random temperatures between 20 and 40
sales = np.random.poisson(lam=30, size=len(days)) + (temperatures - 20) * 2  # Sales influenced by temperature

# Create a DataFrame
df_data = pd.DataFrame({
    'Day': days,
    'Temperature': temperatures,
    'Ice_Cream_Sales': sales
})

# Step 2: Create the interactive scatter plot using Plotly Express
fig = px.scatter(
    df_data,
    x='Temperature',
    y='Ice_Cream_Sales',
    title='Ice Cream Sales vs Temperature',
    labels={'Ice_Cream_Sales': 'Sales (Units)', 'Temperature': 'Temperature (°C)'},
    hover_data=['Day']  # Show the day when hovering over points
)

# Step 3: Show the plot
fig.show()
```

**Exp 9: Interactive Line Plot with Bokeh**


---


In this experiment, we will create an interactive line plot using Bokeh to visualize the daily website traffic over a month. This helps us analyze trends and fluctuations in user visits.

```python
!pip install bokeh
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource

# Step 1: Generate random website traffic data
np.random.seed(42)  # For reproducibility
days = pd.date_range(start='2024-01-01', periods=30).date
traffic = np.random.randint(100, 1000, size=len(days))  # Random traffic numbers

# Create a DataFrame
df_traffic = pd.DataFrame({
    'Day': days,
    'Traffic': traffic
})

# Step 2: Prepare data for Bokeh
source = ColumnDataSource(df_traffic)

# Step 3: Create the interactive line plot
output_notebook()  # Display output in Jupyter Notebook

p = figure(title="Daily Website Traffic Over a Month",
           x_axis_label='Date',
           y_axis_label='Traffic (Visits)',
           x_axis_type='datetime',
           width=800)  # Set plot width

# Create line and scatter plot
p.line('Day', 'Traffic', source=source, line_width=2, color='blue', legend_label='Traffic')
p.scatter('Day', 'Traffic', source=source, size=8, color='blue', alpha=0.6)  # Updated to scatter()

# Step 4: Customize the plot
p.legend.location = "top_left"
p.grid.grid_line_alpha = 0.3

# Step 5: Show the plot
show(p)
```

**Exp 10: Interactive Bar Chart with Bokeh**



---

In this experiment, we will create an interactive bar chart using Bokeh to  visualize the number of different types of fruits sold over a week in a local market.This data can provide insights into consumer preferences and seasonal availability.



```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource

# Step 1: Create sample fruit sales data
data = {
    'Fruit': ['Apples', 'Bananas', 'Cherries', 'Dates', 'Elderberries'],
    'Sales': [150, 200, 120, 80, 90]  # Number of fruits sold
}

df_fruit_sales = pd.DataFrame(data)

# Step 2: Prepare data for Bokeh
source = ColumnDataSource(df_fruit_sales)

# Step 3: Create the interactive bar chart
output_notebook()  # Display output in Jupyter Notebook

p = figure(x_range=df_fruit_sales['Fruit'], title="Fruit Sales in a Local Market Over a Week",
           x_axis_label='Fruit Type', y_axis_label='Number of Fruits Sold',
           width=800, height=400)

# Create the bar chart
p.vbar(x='Fruit', top='Sales', source=source, width=0.9, color='pink')

# Step 4: Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.y_range.end = max(df_fruit_sales['Sales']) + 20  # Extend y-axis for better visibility
p.title.text_font_size = '16pt'
p.xaxis.axis_label_standoff = 12
p.yaxis.axis_label_standoff = 12

# Step 5: Show the plot
show(p)
```

**Exp 11: Interactive Scatter Plot with Bokeh**


---

In this experiment, we will visualize the correlation between the number of hours students studied and the exam scores they achieved. Understanding this relationship can provide insights into the effectiveness of study time on academic performance.


```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool

# Step 1: Create sample study hours and exam scores data
data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva',
                'Frank', 'Grace', 'Hannah', 'Ian', 'Judy'],
    'Hours_Studied': [2, 5, 3, 8, 6, 1, 4, 9, 7, 5],  # Hours studied
    'Exam_Score': [65, 85, 70, 95, 90, 50, 75, 88, 92, 80]  # Corresponding exam scores
}

df_study_scores = pd.DataFrame(data)

# Step 2: Prepare data for Bokeh
source = ColumnDataSource(df_study_scores)

# Step 3: Create the interactive scatter plot
output_notebook()  # Display output in Jupyter Notebook

p = figure(title="Hours Studied vs. Exam Scores",
           x_axis_label='Hours Studied', y_axis_label='Exam Score',
           width=800, height=400)

# Add scatter points
p.scatter('Hours_Studied', 'Exam_Score', source=source, size=10, color='green', alpha=0.6)

# Step 4: Add Hover Tool
hover = HoverTool()
hover.tooltips = [("Student", "@Student"), ("Hours Studied", "@Hours_Studied"), ("Exam Score", "@Exam_Score")]
p.add_tools(hover)

# Step 5: Show the plot
show(p)
```

**Exp 12: Visualizing Clusters with scikit-learn and Plotly**


---


In this experiment, we will perform clustering on a synthetic dataset representing different species of flowers based on their sepal length and sepal width. Using the K-Means clustering algorithm from scikit-learn, we will identify clusters within the data and visualize the results with an interactive Plotly scatter plot. This analysis can help us understand how different species are grouped based on their physical attributes


```python
!pip install scikit-learn plotly
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Step 1: Create synthetic dataset with clusters
n_samples = 300
n_features = 2
n_clusters = 3
random_state = 42

X, y = make_blobs(n_samples=n_samples, centers=n_clusters, n_features=n_features, random_state=random_state)
df_clusters = pd.DataFrame(X, columns=['Sepal_Length', 'Sepal_Width'])

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
df_clusters['Cluster'] = kmeans.fit_predict(X)

# Step 3: Create interactive scatter plot with Plotly
fig = px.scatter(df_clusters, x='Sepal_Length', y='Sepal_Width',
                 color='Cluster', title='K-Means Clustering of Flower Species',
                 labels={'Sepal_Length': 'Sepal Length', 'Sepal_Width': 'Sepal Width'},
                 color_continuous_scale=px.colors.sequential.Viridis)

# Step 4: Show the plot
fig.show()
```
