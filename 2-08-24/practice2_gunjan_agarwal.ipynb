6. Write a Python program to Read a Number n and Print the Sum of odd Natural Numbers between the range of 1 to n both inclusive

```python
n=int(input("Enter a no: ")) #taking input from user
sum=0 #initializing a sum variable to store the summation of odd nos

#using for loop to iterate from 1 to n+1 since both nos are to be inclusive
for i in range(1,n+1):
  #using if statement to check whether the no is odd or not
  if i%2!=0:
    #if no is odd then adding it to sum
   sum=sum+i

#printing the sum
print("Sum of odd nos between 1 and "+str(n)+" is "+ str(sum))

```

7. Write a Python program to find factorial of a number


```python
#taking input from user
n=int(input("Enter a no: "))
#initializing variable to store result
r=1

#using for loop to iterate
for i in range(1,n+1):
  r=r*i

print("Factorial of "+ str(n) + " is " + str(r))
```

8. Write a python program to reverse a given number

```python
#method 1

#taking input from user
n=int(input("Enter a no to reverse: "))
#converting no into string
strn=str(n)
#reversing the string
rno=strn[::-1]
#printing it
print("reversed no is: ", int(rno))
```

```python
#method 2

#taking input from user
n=int(input("Enter a no to reverse: "))
#variable initialized to store reversed no
rno=0

#using while loop
while n>0:
  #getting the last digit
  lastd=n%10
  #adding last digit to the reversed no
  rno=rno*10+lastd
  #removing last digit from original no
  n=n//10

#printing the results
print("Reversed no is: " + str(rno))
```

9. Write a program to print all the strong numbers between 1 to 100
Note: Strong number is a special number whose sum of the factorial of digits is equal to the original number. For Example: 145 is a strong number. Since, 1! + 4! + 5!


```python
#using math module to access factorial function to compute factorial
import math
print("The strong numbers between 1 and 100 are : ")

#using for loop to iterate from 1 to 100
for i in range(1,101):
  sum=0
  n=i
  #using while loop
  while n>0:
  #getting the last digit
   lastd=n%10
   #adding  factorial of digit to the sum
   sum+=math.factorial(lastd)
   #removing last digit from original no
   n=n//10

#checking if the sum is the same as number
  if sum==i:
    print(i)
```

10: Accept the marks for the number of subjects studying in this semester. While accepting marks check the constraints that entered marks should not be negative as well as should not be more than 100 (if entered terminate the code). If the constraint is satisfied, calculate the total & percentage. If % is greater than equal to 92 display “Merit” if % is between 75 and 91 display “Distinction” if % is between 60 and 74 “First class” if %  is between 45 to 59  display “Second class” else display “Fail”



```python
def main():
    # Input number of subjects
    try:
        num_sub = int(input("Enter the number of subjects: "))
        if num_sub <= 0:
            print("Number of subjects should be positive.")
            return
    except ValueError:
        print("Invalid input. Please enter a valid number.")
        return

    t_m = 0
    #using for loop to iterate for getting marks from user
    for i in range(num_sub):
        try:
            m = float(input(f"Enter marks for subject {i + 1}: "))
            #using if loop to check for constraints
            if m < 0 or m > 100:
                print("Marks should be between 0 and 100. Exiting.")
                return
            t_m += m
        except ValueError:
            print("Invalid input. Please enter a valid number.")
            return

    # Calculating the percentage
    p = (t_m / (num_sub * 100)) * 100

    print(f"\nTotal marks obtained: {t_m}")
    print(f"Percentage: {p}%")

    # Displaying the results based on percentage
    print(" ")
    if p >= 92:
        print("Merit")
    elif 75 <= p < 92:
        print("Distinction")
    elif 60 <= p < 75:
        print("First class")
    elif 45 <= p < 60:
        print("Second class")
    else:
        print("Fail")

if __name__ == "__main__":
    main()
```
