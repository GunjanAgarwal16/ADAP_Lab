i. Consider the following scenario:

Input :  

Principle (amount): 1200

Time: 2         

Rate: 5.4

Output : Compound Interest = 133.099243.

Calculate compound interest. Assume your own variables.

```python
#taking input from user
p=float(input("Enter principle amount: "))
t=int(input("Enter time (in years): "))
r=float(input("Enter rate (in percentage): "))

#formula for compound interest
amount=p*(1+pow((r/100),t))
ci=amount-p

#printing the output
print(f"Compound Interest is: {ci:.6f}")
```

j.Write a program to accept 5 numbers from the user using individual statements. Print the sum, average of the numbers. Check if the sum is less than 100 and print the true false value

```python
#taking input from user
a=int(input("Enter first no: "))
b=int(input("Enter second no: "))
c=int(input("Enter third no: "))
d=int(input("Enter fourth no: "))
e=int(input("Enter fifth no: "))

#calculating the sum and average
sum=a+b+c+d+e
avg=sum/5

#printing the output
print("Sum of the numbers is: "+str(sum))
print("Average of the numbers is: "+str(avg))

#checking if the sum is less than 100 using if else loop
if(sum<100):
  print("True")
else:
  print("False")
```
