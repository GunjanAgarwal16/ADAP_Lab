***Sets in Python***

Characteristics:

Unordered: No defined order for elements.
Mutable: Elements can be added or removed after the set is created.
No Duplicates: Each element must be unique.
Unindexed: Elements cannot be accessed by index.
Uses:

Membership Testing: Efficient for testing if an element is in the set.
Eliminating Duplicates: Useful for storing collections of unique items.
Set Operations: Supports mathematical set operations like union, intersection, and difference.

```python
# Create a set of colors
colors = {'red', 'green', 'blue'}

# Add a new color
colors.add('yellow')

# Remove a color
colors.remove('green')

# Union of two sets
primary_colors = {'red', 'blue'}
all_colors = colors.union(primary_colors)
```

***Dictionaries in Python***
Characteristics:

Unordered: No defined order for elements.
Mutable: Elements can be changed after the dictionary is created.
Key-Value Pairs: Stores data in key-value pairs.
No Duplicate Keys: Each key must be unique.
Uses:

Mapping: Suitable for mapping unique keys to values.
Fast Lookups: Provides fast access to values when the key is known.
Data Representation: Ideal for representing structured data.

```python
# Create a dictionary of person details
person = {'name': 'Alice', 'age': 25, 'city': 'New York'}

# Access a value
name = person['name']

# Add a new key-value pair
person['email'] = 'alice@example.com'

# Remove a key-value pair
del person['city']
```

Looping through a Dictionary (using for loop)





```python
colors =	{
  "color1": "Red",
  "color2": "Green",
  "color3": "Blue",
  "color4": "Magenta",
  "color5": "Pink"
  }

# while looping, default return values are keys of the dictionary

for i in colors:
  print(i)
print("\n")

# to return values in the dictionary

for i in colors:
  print(colors[i])

```

# ***Python Functions***

A function is a block of code which only runs when it is called. You can pass data, known as parameters, into a function. A function can return data as a result.

***def*** keyword is used to define a function in python

```python
def func():                       # function definition
  print("This is a function")

func()                           # function call
```

```python
# function with single argument

def func(cname):
  print("Honda " + cname)

func("Amaze")
func("City")
func("Jazz")
```

```python
# function with multiple arguments

def func(company, cname):
  print(company + " " + cname)

func("Honda", "Amaze")
func("Tata", "Nexon")
func("MG", "Hector")
```

```python
# function with arbitrary arguments.
# If the number of arguments is unknown, add a * before the parameter name. The function will receive a tuple of arguments

def func(*cars):
  print("My favorite car is " + cars[2])

func("Honda Amaze", "Tata Nexon", "MG Hector")
```

```python
# function with Default Parameter Value.
# If we call the function without argument, it uses the default value

def func(cname = "BMW"):
  print("I have a " + cname + " car")

func("Honda Amaze")
func("Tata Nexon")
func()
```

```python
# Passing a List as an Argument

def func(cars):
  for i in cars:
    print("I have " + i + " car")

cnames = ["Honda Amaze", "Tata Nexon", "MG Hector"]

func(cnames)
```

Recursion

```python
def fact(n):
   if n == 1:
       return n
   else:
       return n*fact(n-1)

num=input("Enter the number : ")
n = int(num)

# check if the number is negative
if n < 0:
   print("Invalid input")
elif n == 0:
   print("The factorial of 0 is 1")
else:
   print("The factorial of " + num + " is", fact(n))
```
