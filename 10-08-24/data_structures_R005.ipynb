**Sets**

```python
s1={"yellow", "red", "blue", "purple"}
```

```python
s1.add("pink")
print(s1)
```

```python
s2={"green","black"}
s1.update(s2)
print(s1)
```

```python
s1.remove("red")
print(s1)
```

```python
s1.pop()
print(s1)
```

```python
s1.discard("pink")
print(s1)

s1.discard("white")
print(s1)
```

```python
#for count
print(len(s1))
```

```python
#for sort
print(sorted(s1))
```

```python
s3=s1.copy()
print(s3)
```

```python
#for reverse
rl=list(s1)
rl.reverse()
print(rl)
```
