**ARITHMETIC OPERATORS**

```python
a=int(input("enter 1st no: "))
b=int(input("enter 2nd no: "))
print("\nArithmetic Operators \n ")
print("Addition = ", a+b)
print("Subtraction = ", a-b)
print("Multiplication = ", a*b)
print("Division = ", a/b)
print("Modulus = ", a%b)
print("Exponent = ", a**b)
print("Floor Division = ", a//b)
```

**LOGICAL OPERATORS**

```python
a=10
b=20
print("\nLogical Opertors\n")
print("a<b and b>a : ",a<b and b>a) #and operator
print("a>b or a<b : ", a>b or a<b) #or operator
print("not(a<b and b>a) : ",not(a<b and b>a)) #not operator
```

**ASSIGNMENT** **OPERATORS**

```python
a = 10
b = 46
c = a + b

d = +c
e = -d

f = 1      # Initialize f
f *= e

g = 1      # Initialize g
g /= f

h = 1      # Initialize h
h %= g

print("Assignment operators:\n")

print("c = a + b =", c)
print("d = +c =", d)
print("e = -d =", e)
print("f *= e =", f)
print("g /= f =", g)
print("h %= g =", h)
```

**COMPARISON OPERATORS**

```python
a=int(input("enter 1st no: "))
b=int(input("enter 2nd no: "))
print("Comparison operators: ")
print("a==b : ",a==b)
print("a>=b : ",a>=b)
print("a<=b : ",a<=b)
print("a!=b : ",a!=b)
print("a>b : ", a>b)
print("a<b : ", a<b)
```

**BITWISE OPERATORS**

```python
a=6
b=3
print("Bitwise operators: \n")
print("a & b : ", a&b)
print("a | b : ", a|b)
print("a ^ b : ", a^b)
print("a >> 1 : ", a>>1)
print("b << 1 : ", b<<1)
```
