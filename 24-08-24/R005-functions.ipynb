1. Given a year, determine whether it is a leap year. If it is a leap year, return the Boolean True, otherwise return False.

   Input 1990
   
   Output False  

```python
#function to find leap year: if the yr is divisible by 4 check further. if its divisible by 100, it is not a leap yr unless it is also divisible by 400
def is_leap(y):
  return (y % 4 == 0 and y % 100 != 0) or (y % 400 == 0)

#taking i/p from user
y=int(input("Enter a year: "))
print(is_leap(y))
```

6. Define a function  that returns  Reverse of a number.  

```python
#function to reverse the no
def rev(n):
  n=str(n)
  return n[::-1]

# Taking input from the user
n=int(input("Enter a no: "))
rn=rev(n)
print("Reverse of " + str(n)+" is "+rn)
```

7. Write a Python function that checks whether a passed string is a palindrome or not.

```python
#function to check if string is plaindrome
def pal(s):
  return s==s[::-1]

# Taking input from the user
s=input("Enter a string: ")
if  pal(s):
  print("The string is a palindrome")
else:
  print("The string is not a palindrome")
```
