1. Write a function division() that accepts two arguments. The function should be able to catch an exception such as ZeroDivisionError, ValueError, or any unknown error you might come across when you are doing a division operation.

```python
def divide_numbers(a, b):
    try:
        result = a / b
        print("The result is:", result)
    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")


# Test the function
num1 = int(input("ENter no 1: "))

num2 = int(input("ENter no 2: "))

divide_numbers(num1, num2)
```

```python
def div():
  try:
    n1=float(input("Enter first number: "))
    n2=float(input("Enter second number: "))
    r=n1/n2
    print(f"Division: {r}")
  except ZeroDivisionError:
    print("Division by zero is not allowed")
  except ValueError:
    print("Invalid Input")
  except Exception as e:
    print("Unexpected error")

div()
```

02. Write a Python program that calculates sum of square of n integers provided by the user. The program must raise an exception if user enters data of wrong data type.

```python
def sum(n):
  t=0
  for i in range(n):
    try:
      no=int(input(f"Enter integer {i+1}: "))
      t+=no**2
    except ValueError:
      print("Invalid Input")
  return t

try:
  n=int(input("No of integers to be entered: "))
  print("Sum of squares of nos entered: ",sum(n))
except ValueError:
  print("Invalid Input")
```

3. Write a Python program that prompts the user to input two numbers and raises a TypeError exception if the inputs are not numerical.

```python
def no(prompt):
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Input must be a number.")

def main():
    try:
        n1 = no("Enter the first number: ")
        n2 = no("Enter the second number: ")
    except TypeError:
        print("Both inputs must be numbers.")
    else:
        print("You entered: "+str( n1)+ " and "+str( n2))

if __name__ == "__main__":
    main()
```

4. Write a Python program that executes an operation on a list and handles an IndexError exception if the index is out of range.

```python
l = [10, 45, 32, 21, 23, 6]

try:
    i = int(input("Enter index: "))
    print(l[i])
except IndexError:
    print("Index out of range")
except ValueError:
    print("Invalid input: Please enter an integer.")
```

5. Write a python program to create a method that takes an integer as a parameter and throws an exception if the number is odd.

```python
try:
    n = int(input("Enter a number: "))
    if n % 2 != 0:
        raise Exception("You entered an odd number. Please enter an even number.")
    else:
        print(f"{n} is an even number.")
except Exception as e:
    print(e)
```
